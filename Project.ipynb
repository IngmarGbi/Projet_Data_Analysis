# Random Forest

```python
import scipy.stats as stats
import pylab
import matplotlib.pyplot as plot
```

```python
def RandomForest_grid_search(df, Y):
    """
    Rglage des hyper-paramtres du classifieur Random Forest
    """
    import sys, pickle
    from sklearn.ensemble          import RandomForestClassifier
    from sklearn.model_selection   import GridSearchCV
    from sklearn.metrics           import make_scorer
    from sklearn.preprocessing     import label_binarize
    from sklearn.multiclass        import OneVsRestClassifier
    #grid search :

    parameters = {  "n_estimators"         : [ 10, 100, 200 ], 
                    "min_samples_leaf"     : [  3,   1,   7 ],
                    "criterion"            : ["gini"],#["gini", "entropy"]         ,
                    "max_features"         : ["auto"]      ,
                    "max_depth"            : [None]         , 
                    "n_jobs"               : [ -1]             ,  
                    "class_weight"         : ["auto"]          ,
                    "warm_start"           : [False]           }
    parameters = {  "n_estimators"         : [ 10], 
                    "min_samples_leaf"     : [  3 ],
                    "criterion"            : ["gini"],#["gini", "entropy"]         ,
                    "max_features"         : ["auto"]      ,
                    "max_depth"            : [None]         , 
                    "n_jobs"               : [ -1]             ,  
                    "class_weight"         : ["auto"]          ,
                    "warm_start"           : [False]           }
    
    grid_type = "RandomForestClassifier"
    
    print ("-"*30, "\n %s"%grid_type); sys.stdout.flush()
    """
        Search 
    """
    
    # Exhaustive Grid Search : will try all parameters combinations
    # vs RandomizedSeachCV where the model test random combinations.
    
    grid         = GridSearchCV(RandomForestClassifier(), {})#, parameters, verbose=1, scoring='f1_weighted')
    
    try : grid.fit(df, Y)
    except Exception as e : print ("[%s - 2] : %s"%(grid_type,e) )
    print (" best_score_ %s with =  %s "%( grid.best_score_,  grid.best_estimator_ ) ); sys.stdout.flush()    
    return grid.best_estimator_
```

```python
def print_time():
    """
    Affiche et retourne l'heure de la machine.
    Utilis pour mesurer le temps d'execution du classifieur.
    """
    import time
    import datetime
    t = datetime.datetime.now().time()
    print(t.isoformat()[:8])
    return time.time()
print_time()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(Data.drop('Classification', axis=1), Data["Classification"])
```

```python
#200   rows : 0.26s / 0.68 pred / 0.66 acc
#2000  rows : 1.55s / 0.74 pred / 0.68 acc
#20000 rows : 15.5s / 0.78 pred / 0.72 acc
#nb_lignes = 2000
start_time = print_time()

# Apprentissage du classifieur RF sur X_train et Y_train (<nb_lignes> donnes)
rf = RandomForest_grid_search(X_train, list(Y_train))

end_time = print_time()
print ("\nDuration time :", end_time - start_time)

# REMARQUES :
# L'utilisation du classifieur RF semble appropri pour faire une prdiction sur les donnes que l'on possde.
# L'entrainement du classifieur ncessite 15 secondes pour prs de 20 000 donnes semble tout  fait raisonnable.

```

```python
def draw_confusion_matrix(y_test, predictions, title):
    """
    Cr et affiche la matrice de confusion des donnes de Y_test et de celles sur lesquelles on a fait la prdiction.
    Permet d'valuer la qualit de la classification souhaite.
    Plus les valeurs diagonales sont leves, plus le classifieur prdit correctement.
    """
    from sklearn.metrics import confusion_matrix,accuracy_score
    import matplotlib.pyplot as plt
    cm    = confusion_matrix(y_test, predictions)
    accur = accuracy_score  (y_test, predictions)
    
    plt.matshow  (cm)
    plt.ylabel   ('True Label')
    plt.xlabel   ('Predicted Label')
    plt.suptitle ('Confusion matrix (Accuracy of %.2f) for [%s]'%(accur,title))
    plt.colorbar ()
    plt.show     ()

# Matrice de confusion des donnes X_test sur lesquelles on fait la prdiction et des donnes Y_test.
# La prdiction du modle tant de 0.72, au mieux, commence  tre correct.
# Un grand nombre de Vrais Positifs et Vrais Ngatifs sont prsents.
draw_confusion_matrix(list(Y_test), rf.predict(X_test), "Dtection Classification du sol")

```
